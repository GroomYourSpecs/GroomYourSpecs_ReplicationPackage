```r
install.packages('readxl')
install.packages('likert')
```

```r
library(readxl)
library(likert)
```

```r
bottom_up <- read.csv("/content/sample_data/bottom_up.csv")
View(bottom_up)
```

```r
bottom_up[,3:10] <- lapply(bottom_up[,3:10],
                            factor,
                            levels=1:7,
                            labels = c("Strongly disagree",
                                       "Disagree",
                                       "Somewhat disagree",
                                       "Neutral",
                                       "Somewhat agree",
                                       "Agree",
                                       "Strongly agree"),
                                       ordered=TRUE)

lik <- likert(as.data.frame(bottom_up[ , 3:10]))

title <- "Bottom-up Specification"

plot(lik, wrap = 60, text.size=3, group.order = colnames(bottom_up)[3:10]) + ggtitle(title)
```

```r
top_down <- read.csv("/content/sample_data/top_down.csv")
View(top_down)
```

```r
top_down[,3:10] <- lapply(top_down[,3:10],
                            factor,
                            levels=1:7,
                            labels = c("Strongly disagree",
                                       "Disagree",
                                       "Somewhat disagree",
                                       "Neutral",
                                       "Somewhat agree",
                                       "Agree",
                                       "Strongly agree"),
                            ordered=TRUE,)

lik <- likert(as.data.frame(top_down[,3:10]))

title <- "Top-down Specification"

plot(lik, wrap = 60, text.size=3, group.order = colnames(top_down)[3:10]) + ggtitle(title)
```
